```python
import re
mystring='Python is cool !! '
if re.search('Python',mystring):
    print 'Python found!'
```

```python
word_freq={}
for tok in mystring.split():
    if tok in word_freq:
        word_freq[tok]+=1
    else:
        word_freq[tok]=1
print word_freq
```

```python
import urllib2

response=urllib2.urlopen('http://www.uol.com.br')
html=response.read()
print len(html)
```

```python
tokens=[tok for tok in html.split()]
print "Total de tokens:"+str(len(tokens))
print tokens[0:100]
```

```python
import re
tokens=re.split('\W+',html)
print len(tokens)
print tokens[0:100]
```

```python
import nltk
from bs4 import BeautifulSoup

bts4=BeautifulSoup(html,'html.parser')
clean=bts4.get_text()
tokens=[tok for tok in clean.split()]
print tokens[:100]
```

```python
import operator
freq_dist={}
for tok in tokens:
    if tok in freq_dist:
        freq_dist[tok]+=1
    else:
        freq_dist[tok]=1
sorted_freq_dist=sorted(freq_dist.items(),key=operator.itemgetter(1),reverse=True)
print sorted_freq_dist
```

```python
import nltk
Freq_dist_nltk=nltk.FreqDist(tokens)
print Freq_dist_nltk
for k,v in Freq_dist_nltk.items():
    try:
        print str(k).decode('UTF-8')+':'+str(v).decode('UTF-8')
    except:
        print('Caracter nao reconhecido')
    
    
```

```python
Freq_dist_nltk.plot(50,cumulative=False)
```
